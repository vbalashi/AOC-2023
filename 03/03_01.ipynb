```python
import pandas as pd
from param import DataFrame
import re

task = """467..114..
...*......
..35..633.
......#...
617#......
.....+.58.
..592.....
......755.
...$.*....
.664.598..
"""
lines = task.splitlines()
# df = pd.DataFrame({'line': lines})
df = pd.read_csv('03_task02.txt', names=['line'])

pattern = "\d+"
part_numbers = []

def process_line(df: DataFrame, line_id: int):
    line = df.iloc[line_id]['line']
    matches = re.finditer(pattern, line)
    
    previous_line_id = line_id
    if line_id > 0:
        previous_line_id = line_id - 1
    next_line_id = line_id
    if line_id < len(df) - 1:
        next_line_id = line_id + 1

    lines_to_check = range(previous_line_id, next_line_id+1)

    for match in matches:
        #finding all indexes of possible locations for our match
        first_position = 0
        if match.start() > 0:
            first_position = match.start() - 1
        last_position = match.end() - 1
        if match.end() < len(line) - 1:
            last_position = match.end()
        
        positions = range(first_position, last_position+1)
        # print(f"{line_id}: {line} {lines_to_check} {positions}")


        def get_character(df, line_number, char_index):
            line = df.iloc[line_number]['line']
            return line[char_index]

        for line_id_being_checked in lines_to_check:
            exit_outer_loop = 0
            # print(f"{line_id_being_checked} {df.iloc[line_id_being_checked]['line']}")
            for char_id in positions:
                # print(f"{line_id}: {char_id}")
                char = get_character(df, line_id_being_checked, char_id)
                # print(f"{line}   line_id:{line_id}   line_id_being_checked:{line_id_being_checked}   char:{char_id} is {char}   match.start:{match.start()}, match.end {match.end()}")
                if not(line_id == line_id_being_checked and char_id in range(match.start(), match.end())):
                    if char != "." and not char.isdigit(): 
                        # print(f"{line_id}: {line} {lines_to_check} {positions} it's a match! line: {line_id_being_checked} char: {char_id}")
                        part_numbers.append(match.group(0))
                        exit_outer_loop = 1
                        break
            if exit_outer_loop == 1:
                break

for line_id in range(len(df)):
    process_line(df, line_id)

part_numbers_int = [int(x) for x in part_numbers]
print(part_numbers_int)
print(sum(part_numbers_int))
```
